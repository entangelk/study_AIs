```python
import pandas as pd
```

```python
import pyLDAvis
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
# 데이터 로드
data = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
documents = data.data
```

```python
documents[:2]
```

### TfidfVectorizer options
반복되는것과 반복되지 않는것을 거르는데 사용 - 파라미터 -> 소수(확률), 정수(갯수)
- max_df : 이상이면 빼버려라 (너무 많아서 쓸모 없을 때)
- min_df : 이하면 빼버려라 (너무 적어서 쓸모 없을 때)

```python
tfidfVectorizer = TfidfVectorizer(stop_words='english',max_df=0.95, min_df=2)
```

```python
tfidf = tfidfVectorizer.fit_transform(documents)
tfidf.toarray()[:2]
```

```python
import numpy as np
np.where(tfidf.toarray()[:2] > 0)
```

```python
latentDirichletAllocation = LatentDirichletAllocation(n_components=5,n_jobs=-1)
```

```python
latentDirichletAllocation_seven = LatentDirichletAllocation(n_components=7,n_jobs=-1)
```

```python
latentDirichletAllocation_three = LatentDirichletAllocation(n_components=3,n_jobs=-1)
```

```python
latentDirichletAllocation.fit(tfidf)
```

```python
latentDirichletAllocation_three.fit(tfidf)
```

```python
latentDirichletAllocation_seven.fit(tfidf)
```

```python
vis_five = pyLDAvis.lda_model.prepare(latentDirichletAllocation, tfidf, tfidfVectorizer)
vis_seven = pyLDAvis.lda_model.prepare(latentDirichletAllocation_seven, tfidf, tfidfVectorizer)
vis_three = pyLDAvis.lda_model.prepare(latentDirichletAllocation_three, tfidf, tfidfVectorizer)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis_five)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis_seven)
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.display(vis_three)
```
