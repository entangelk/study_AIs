```python
import pandas as pd
```

```python
df_car = pd.read_csv('LetalCarOfContractType.csv')
df_car
```

```python
df_car.columns
```

### 데이터 전처리(Pre-processing)
- 학습 모델을 만들기 위해 이상치, 결측치 drop
- age와 mileage는 동일하게 움직이므로 mileage는 학습 데이터에서 제외
- 수치형 데이터 규합 (term, amount,overdue_count,credit rating)

```python
leandata = df_car.dropna().copy()
leandata
```

```python
leandata['Term'].unique()
```

```python
leandata['amount'].unique()
```

```python
leandata['overdue_count'].unique()
# -1은 있을 수 없는 숫자이므로 제외
```

```python
leandata = leandata.query('overdue_count != -1')
```

```python
leandata['overdue_count'].unique()
```

```python
leandata['credit rating'].unique()
```

### 데이터 분리
- target과 labels(features) 분리
- train과 test set 분리

```python
target = leandata['age']
features = leandata[['Term', 'amount', 'overdue_count','credit rating']]
target.shape, features.shape
```

```python
from sklearn.model_selection import train_test_split
# X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=321)
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.3, random_state=55)
features_train.shape, features_test.shape, target_train.shape, target_test.shape
```

## 모델

### 목표 변수에 따른 모델 선택 - 연속형
- 목표 변수
- 설병 변수

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# 학습 진행
model.fit(features_train,target_train)
```

## 평가

```python
from sklearn.metrics import r2_score
```

```python
# train data set
target_train_predict = model.predict(features_train)
target_train.shape, target_train_predict.shape # 원래 정답과 학습 후 정답
```

```python
# train data set에 대한 평가
r2_score(target_train, target_train_predict)
```

```python
target_test_predict = model.predict(features_test)
target_test.shape, target_test_predict.shape # 테스트 그룹 확인
```

```python
r2_score(target_test, target_test_predict)
```

## 서비스


```python
# 재 사용 위해 model을 파일로 저장
import pickle # 메모리 인스턴스 클래스를 이진파일로 저장
```
