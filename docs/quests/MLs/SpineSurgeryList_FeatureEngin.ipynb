```python
import pandas as pd
```

```python
get_data = pd.read_csv('SpineSurgeryList.csv')
get_data
```

```python
get_data.isna().sum()
```

## 목표변수, 설명변수 선택
- 목표변수 : 재발여부
- 설명변수 :
  + 수치형 : 연령, 체중
  + 범주형 : 고혈압여부, 당뇨여부, 빈혈여부, 성별, 흡연여부

## 데이터 전처리
- 이상치 제거 (3표준편차)
- 결측치 없는 데이터들만 사용

```python
fare_mean = get_data['체중'].mean()
fare_std = get_data['체중'].std()
get_data = get_data.query(f'체중 < {fare_mean} + {fare_std}*3 & 체중 > {fare_mean} - {fare_std}*3')
```

```python
fare_mean = get_data['연령'].mean()
fare_std = get_data['연령'].std()
get_data = get_data.query(f'연령 < {fare_mean} + {fare_std}*3 & 연령 > {fare_mean} - {fare_std}*3')
```

```python
get_data
```

```python
df_SSL_extract = get_data[['재발여부','연령', '체중','고혈압여부', '당뇨여부', '빈혈여부', '성별', '흡연여부']]
df_SSL_extract.isna().sum()
```

```python
df_SSL_extract = df_SSL_extract.reset_index(drop=True) # 인덱스 초기화
```

## 특성공학(Feature Engineering)
- 수치 평준화 : 수치형과 범주형 각각 적용
- 갯수 평준화 : target 수량

### 스케일링 scaling - 수치형에 적용
- Standard Scalin : 평균 0, 표준편차 1
- Min-Max Scaling : 0 ~ 1 사이 (최소 0 ~ 최대 1)
- Robust Scaling : 이상치가 많은 데이터 셋(중앙값 기준)

```python
# ['연령', '체중']
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
```

```python
minmax.fit(df_SSL_extract[['연령']])
df_SSL_extract['연령_scaler'] = minmax.transform(df_SSL_extract[['연령']])
```

```python
minmax.fit(df_SSL_extract[['체중']])
df_SSL_extract['체중_scaler'] = minmax.transform(df_SSL_extract[['체중']])
```

### Onehot encoding은 범주형(features)에 적용

```python
# ['고혈압여부', '당뇨여부', '빈혈여부', '성별', '흡연여부']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
def onehot(get_list,df):
  for i in get_list:
    ohe_list = OneHotEncoder()
    ohe_list.fit(df[[i]]) # 학습진행
    encoder_list = ohe_list.transform(df[[i]]).toarray()
    df_encoder_list = pd.DataFrame(data=encoder_list, columns=ohe_list.get_feature_names_out())
    df = pd.concat((df,df_encoder_list), axis=1)
    df = df.drop(columns=i)
  return df
```

```python
df_SSL_extract = onehot(['고혈압여부', '당뇨여부', '빈혈여부', '성별', '흡연여부'],df_SSL_extract)
```

```python
df_SSL_extract
```

### 정형화 단계 - target과 feature 분리

```python
target_train = df_SSL_extract['재발여부']
features_train = df_SSL_extract.drop(columns=['재발여부'])
target_train.shape, features_train.shape
```

## 모델

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(features_train,target_train)
```

## 평가

```python
model.predict(features_train[10:14])
```

```python
model.predict_proba(features_train[10:14])
```

### F1 score
- 정밀도, 재현율 두가지를 같이 볼 수 있게 해주는 수치 : F1

요걸 좀 더 많이 씀

```python
from sklearn.metrics import classification_report
target_train_prdict = model.predict(features_train)
print(classification_report(target_train,target_train_prdict)) # 그냥하면 보기 힘들어서 프린트 해줘야댐
# f1 스코어를 봐야함
# 0 1 일때 f1스코어가 비슷하지 않는다 => 치우침이 있다
# macro avg = 일반적인 평균
# weighted avg = 가중치가 있는 평균 (쏠림이 있을때 가중치를 줘서 평균을 줌)
```
